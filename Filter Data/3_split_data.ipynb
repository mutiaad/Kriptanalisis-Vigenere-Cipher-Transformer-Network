## Plaintext Validasi

```python
### plain-validasi

import pandas as pd
import json

# baca excel
df = pd.read_excel("Vigenere.xlsx")

# ambil kolom pertama, baris 1-1250
data = df.iloc[:1250, 0]

# pastikan semua jadi string
data = data.astype(str).tolist()

# simpan ke json
with open ("plain-val.json", "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=2)
```

## Plaintext Training

```python
### plain-training

import pandas as pd
import json

# baca excel
df = pd.read_excel("Vigenere.xlsx")

# ambil kolom pertama, baris 1251-5000
data = df.iloc[1250:5000, 0]

# pastiin semua jadi string
data = data.astype(str).tolist()

# simpan ke json
with open ("plain-train.json", "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=2)
```
