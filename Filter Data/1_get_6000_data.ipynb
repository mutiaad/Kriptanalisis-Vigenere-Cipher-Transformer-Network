```python
import pandas as pd
import re

# File
input_file = "rockyou.txt"
output_csv = "filtered_passwords.csv"
output_xlsx = "filtered_passwords.xlsx" 
max_data = 6000

# Pola regex: hanya huruf besar, huruf kecil, dan angka, panjang 8-12
pattern = re.compile(r'^[A-Za-z0-9]{8,12}$')

def read_and_filter_passwords(file_path, max_count):
    valid_passwords = []
    try:
        with open(file_path, "r", encoding="utf-8", errors="ignore") as file:
            for line in file:
                pwd = line.strip()
                if pattern.match(pwd):
                    valid_passwords.append(pwd)
                    if len(valid_passwords) >= max_count:
                        break
    except FileNotFoundError:
        print(f"File '{file_path}' tidak ditemukan.")
        return []
    except Exception as e:
        print(f"Terjadi kesalahan saat membaca file: {e}")
        return []
    return valid_passwords

# Proses filter
password_list = read_and_filter_passwords(input_file, max_data)

if password_list:
    # Simpan ke CSV dan Excel
    df = pd.DataFrame(password_list, columns=["Password"])
    df.to_csv(output_csv, index=False, encoding="utf-8")
    df.to_excel(output_xlsx, index=False)

    print(f"Berhasil memproses {len(password_list)} password.")
    print(f"Disimpan ke: {output_csv} dan {output_xlsx}")
else:
    print("Tidak ada password yang memenuhi kriteria.")
```
